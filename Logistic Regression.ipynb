```python
# Import necessary libraries
import numpy as np
import pandas as pd
from sklearn.utils import shuffle
from sklearn.datasets import fetch_openml

mnist = fetch_openml('mnist_784')

# Convert the data to a DataFrame
df = pd.DataFrame(mnist.data, columns=mnist.feature_names)

# Add the target column to the DataFrame
df['target'] = pd.Series(mnist.target)

# Subset the data to include only class 0 and class 1
df = df[df['target'].isin(['0', '1'])]


```

#### Standardize

```python
def standradize(data):
  return (data - data.mean()) / data.std()
```

```python
## Calculate accuracy
def accuracy(y, y_hat):
    true=0
    for i in range(len(y)) :
        if y[i] == y_hat[i]:
            true+=1
    return (true / len(y)) * 100
```

#### Implement Logistic Regression

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

```python
def cost(y,y_hat):
    a=-y * np.log(y_hat)
    b= ((1-y) * np.log(1-y_hat))
    return np.mean(a-b)
```

```python
def fit(X, y, eta, epochs):
  w = np.random.rand(len(X[0]))
  b = 0
  for epoch in range(epochs):
    z = np.dot(X,w)+b
    z = sigmoid(z)
    y_pred = z.copy()
    #calculate error
    error = cost(y,y_pred)
    # update weights and bias
    w = w - ((eta * np.dot((y_pred - y),X)) / len(X))
    b = b -( eta * np.mean((y_pred - y)))
    if error ==0:
      break
  return w,b      
```

```python
def unit_step(z):
        for i in range(len(z)):
            if z[i] >0.5:
                z[i] = 1
            else:
                z[i]=0
        return z
```

```python
def predict(X,w,b):
        return unit_step((np.dot(w, X.T)) + b)
```

#### Divide data into training and validation set using 10-fold cross validation method

```python
def kFold_cross_validation(X, y, k, eta, epochs):
  
    
  fold = len(X) //k
  accuacy=[]
  for i in range(k):
        start = i * fold
        end= (i + 1) * fold
        # validation data
        X_val, y_val = X[start:end], y[start:end]
        # train data
        X_train = np.concatenate((X[:start], X[end:]), axis=0)
        y_train = np.concatenate((y[:start], y[end:]), axis=0)
        standradize(X_train)
        # train model 
        w,b=fit(X_train, y_train, eta, epochs)
        y_hat = predict(X_val,w.T,b)
        m = len(y_hat)
        # print accuracy 
        print("Accuracy of ", i+1, "is : ", accuracy(y_val, y_hat))
        print("--------------------------------------------------------------")
        accuacy.append(accuracy(y_val, y_hat))
  avg=0     
  for i in accuacy:
    avg+=i
  print(f"the average of accuracy is : {avg/k}")
```
